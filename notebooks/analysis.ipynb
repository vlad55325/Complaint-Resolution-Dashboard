# Complaint Resolution Analysis

Exploratory analysis of complaint dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/sample_complaints.csv')
df.head()
```

```python
# Complaint counts per category
category_counts = df['Category'].value_counts()
sns.countplot(data=df, x='Category', order=category_counts.index)
plt.title('Number of complaints per category')
plt.xticks(rotation=45)
plt.show()
```
